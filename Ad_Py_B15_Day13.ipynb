```python
from matplotlib import pyplot as plt
```

```python
#To give graphical presentation of the data.
```

```python
x = [1,2,3]
y = [1,4,6]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.title("sales indication")
plt.xlabel("sales per year")
plt.ylabel("revenue pr year")
plt.plot(x,y)
plt.show()
```

```python
#Business Intelligence(BI)
```

```python
z = [10,5,0]
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
plt.title("sales indication")
plt.xlabel("sales per year")
plt.ylabel("revenue per year")
plt.legend(["this is x and y","this is x and z"])
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```
